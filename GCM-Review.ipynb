```python
import intake
import pandas as pd

url = "https://storage.googleapis.com/cmip6/pangeo-cmip6.json"
col = intake.open_esm_datastore(url)
df_info = col.df
grouped_info = df_info.groupby(['source_id', 'experiment_id'])['variable_id'].unique().reset_index()
result_df = pd.DataFrame(grouped_info)
result_df.columns = ['Model', 'Scenario','Variables']
```

```python
result_df
```

```python
var_id = result_df[result_df['Variables'].apply(lambda x: all(v in x for v in ['pr', 'tas', 'hur']))].reset_index()
```

```python
var_id = var_id.drop("index", axis = 1)
var_id
```

```python
var_id = var_id.pivot(index='Model', columns='Scenario', values='Variables').reset_index()
var_id.head()
```

```python
exp_id = var_id[['Model','ssp119','ssp126', 'ssp245', 'ssp370', 'ssp434', 'ssp460','ssp585']]
```

```python
print(tabulate(exp_id.head(), headers='keys', tablefmt='pretty'))
```

```python
exp_id = exp_id.dropna(subset=exp_id.columns[1:], how='all')
```

```python
import numpy as np

df = exp_id.copy()
df = df.apply(lambda col: col.apply(lambda x: x.tolist() if isinstance(x, np.ndarray) else x))
```

```python
def select_specific_words(word_list):
    if isinstance(word_list, list):
        return [word for word in word_list if word in ['tas', 'pr', 'hur']]
    else:
        return []
df.iloc[:, 1:] = df.iloc[:, 1:].applymap(select_specific_words)
```

```python
df = df.applymap(lambda x: " " if isinstance(x, list) and not x else x)
```

```python
print(tabulate(df, headers='keys', tablefmt='psql'))
```
